```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data_complete['school_name'].unique())
total_students = len(school_data_complete['Student ID'].unique())
total_budget = sum(school_data_complete['budget'].unique())
avg_math_score = school_data_complete['math_score'].mean()
avg_reading_score = school_data_complete['reading_score'].mean()
pass_math = school_data_complete[school_data_complete.math_score > 69]
perc_math = (len(pass_math['math_score'])/len(school_data_complete['math_score'])) * 100
pass_read = school_data_complete[school_data_complete.reading_score > 69]
perc_read = (len(pass_read['reading_score'])/len(school_data_complete['reading_score'])) * 100
perc_pass = ((perc_math + perc_read)/total_students) * 100
summary_df = pd.DataFrame(((total_schools, total_students, total_budget, avg_math_score, avg_reading_score, perc_math, perc_read, perc_pass),),columns=['Total Schools','Total Students','Total Budget','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing'])
summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
names = school_data_complete['school_name'].unique()
indexed_names = school_data_complete.set_index('school_name')
school_data = list()
for name in names:
    filtered_school_data = indexed_names.loc[name]
    school_data.append([
        name,
        filtered_school_data['type'][0],
        len(filtered_school_data['Student ID'].unique()),
        filtered_school_data['budget'][0],
        (filtered_school_data['budget'][0])/(len(filtered_school_data['Student ID'].unique())),
        filtered_school_data['math_score'].mean(),
        filtered_school_data['reading_score'].mean(),
        len(filtered_school_data[filtered_school_data.math_score > 69]['math_score'])/len(filtered_school_data['math_score']) * 100,
        len(filtered_school_data[filtered_school_data.math_score > 69]['reading_score'])/len(filtered_school_data['reading_score']) * 100,
        (len(filtered_school_data[filtered_school_data.math_score > 69]['math_score'])/len(filtered_school_data['math_score']) * 100)/(len(filtered_school_data['Student ID'].unique())) * 100
    ])
    #print(filtered_school_data)   
#school_data
#dataframe
schoolsummary_df = pd.DataFrame(school_data,columns=['','School Type','Total Students','Total Budget','Per Student Budget','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing'])
schoolsummary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
col = "math_score"
school_grade_data = []
for name in names:
    filtered_school_data = indexed_names.loc[name]
    filtered_school_data = filtered_school_data.loc[:, ["grade", col]] # Remove every column except for the ones we want
    grade_data = [name]
    grades = filtered_school_data["grade"].unique()
    for grade in grades:
        filtered_grade_data = filtered_school_data.set_index("grade").loc[grade]
        grade_data.append(round(filtered_grade_data[col].mean(),2))
    school_grade_data.append(grade_data)

school_summarydf = pd.DataFrame(school_grade_data, columns=['', *grades]).set_index('').sort_values("")
school_summarydf.index.name = None # Remove the space under the column names
school_summarydf
```
